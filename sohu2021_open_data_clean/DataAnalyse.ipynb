# 数据分析

## 文本长度分布

```python
import json
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 短短匹配
ss_a_path = "短短匹配A类/"
ss_b_path = "短短匹配B类/"
# 短长匹配
sl_a_path = "短长匹配A类/"
sl_b_path = "短长匹配B类/"
# 长长匹配
ll_a_path = "长长匹配A类/"
ll_b_path = "长长匹配B类/"
```

```python
class Inputfeature():
    def __init__(self, source, target, label):
        self.s = source
        self.t = target
        self.l = label
        
def get_data(path, label_type):
    features = []
    files = os.listdir(path)
    for file in files:
        if "train" in file or "valid" in file:
            with open(path+file, "r") as f:
                for line in f.readlines():
                    raw = json.loads(line)
                    s, t, l = raw.keys()
                    source = raw[s]
                    target = raw[t]
                    label = int(raw[l])
                    features.append(Inputfeature(source, target, label))
    return features
```

```python
ss_a_feats = get_data(ss_a_path, "A")
ss_b_feats = get_data(ss_b_path, "B")
sl_a_feats = get_data(sl_a_path, "A")
sl_b_feats = get_data(sl_b_path, "B")
ll_a_feats = get_data(ll_a_path, "A")
ll_b_feats = get_data(ll_b_path, "B")
```

```python
# 获取用于长度分布绘图的长度数据列表
def get_length_data(feats):
    all_a_t = [d.t for d in feats]
    all_a_s = [d.s for d in feats]
    
    print(len(all_a_t), len(all_a_s))
    
    a_t_lengths = [len(d) for d in all_a_t]
    a_s_lengths = [len(d) for d in all_a_s]
    
    return a_t_lengths, a_s_lengths
```

```python
# 绘制长度分布
def plotLengthDistribution(s_length, t_length):
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    ax[0].hist(s_length, bins=40, label="Distribution of Source")
    ax[0].set(xlabel="length of sentence", ylabel="Nums")
    ax[0].legend()
    ax[1].hist(t_length, bins=40, label="Distribution of Target")
    ax[1].set(xlabel="length of sentence", ylabel="Nums")
    ax[1].legend()
```

### 短短文本数据长度分布

```python
ss_a_t_lengths, ss_a_s_lengths = get_length_data(ss_a_feats)
plotLengthDistribution(ss_a_s_lengths, ss_a_t_lengths)
```

### 短长文本数据长度分布

```python
all_sl_a_t, all_sl_a_s = get_length_data(sl_a_feats)
plotLengthDistribution(all_sl_a_s, all_sl_a_t)
```

### 长长文本数据长度分布

```python
all_ll_a_t, all_ll_a_s = get_length_data(ll_a_feats)
plotLengthDistribution(all_sl_a_s, all_sl_a_t)
```

### 0/1 标签分布

```python
def get_label_data(feats):    
    labels = [d.l for d in feats]
    
    return labels
```

```python
ss_a_labels = get_label_data(ss_a_feats)
ss_b_labels = get_label_data(ss_b_feats)
sl_a_labels = get_label_data(sl_a_feats)
sl_b_labels = get_label_data(sl_b_feats)
ll_a_labels = get_label_data(ll_a_feats)
ll_b_labels = get_label_data(ll_b_feats)
```

```python
def getDataForLabelDist(labels):
    new = [labels.count(0), labels.count(1)]
    return new
```

```python
ss_a = getDataForLabelDist(ss_a_labels)
ss_b = getDataForLabelDist(ss_b_labels)
sl_a = getDataForLabelDist(sl_a_labels)
sl_b = getDataForLabelDist(sl_b_labels)
ll_a = getDataForLabelDist(ll_a_labels)
ll_b = getDataForLabelDist(ll_b_labels)
```

```python
fig, ax = plt.subplots(3, 2, figsize=(12, 12))
x = ["0", "1"]
ax[0][0].bar(x, ss_a, label="ShortShort label A")
ax[0][0].set(xlabel="Categories", ylabel="Nums")
ax[0][0].legend()
ax[0][0].set_xticks(x, ["0", "1"])
ax[0][1].bar(x, ss_b, label="ShortShort label B")
ax[0][1].set(xlabel="Categories", ylabel="Nums")
ax[0][1].legend()
ax[1][0].bar(x, sl_a, label="ShortLong label B")
ax[1][0].set(xlabel="Categories", ylabel="Nums")
ax[1][0].legend()
ax[1][1].bar(x, sl_b, label="ShortLong label B")
ax[1][1].set(xlabel="Categories", ylabel="Nums")
ax[1][1].legend()
ax[2][0].bar(x, ll_a, label="LongLong label B")
ax[2][0].set(xlabel="Categories", ylabel="Nums")
ax[2][0].legend()
ax[2][1].bar(x, ll_b, label="LongLong label B")
ax[2][1].set(xlabel="Categories", ylabel="Nums")
ax[2][1].legend()
```

## 总结

基于上述的分布，我们大致可以尝试将数据分为两部分进行训练，其中【短短文本】单独训练，【短长】和【长长】可以放在一起训练。

此外，部分类别中，正负分布不均，若训练效果不好，后期可进行数据增强，改善0/1类别平衡性。
